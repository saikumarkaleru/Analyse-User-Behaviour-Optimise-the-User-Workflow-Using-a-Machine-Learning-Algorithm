```python
import seaborn as sns
from markovclick import dummy
from markovclick.models import MarkovClickstream
```

```python
import pandas as pd
df=pd.read_csv('C:/Users/welcome/Desktop/MLprojectcsv.csv',header=None)
```

```python
data=[]
for i in range(74):
    x=df.values[i]
    x=x[~pd.isnull(x)]
    data.append(x[1:])
```

```python
m = MarkovClickstream(data)

s=sns.heatmap(m.prob_matrix, xticklabels=m.pages, yticklabels=m.pages,cmap="YlGnBu",annot=True)
```

```python
from markovclick.viz import visualise_markov_chain
graph = visualise_markov_chain(m)
```
